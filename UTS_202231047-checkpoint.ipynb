## Import Library

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

## Membaca Data Gambar

```python
img = cv2.imread("Dewi.jpg")
```

```python
img.shape
```

## Membuat Baris Dan Kolom

```python
(baris,kolom) = img.shape[:2]
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

## Membuat Citra Kontras


```python
alpa = 1.15
citra_kontras = np.zeros((baris, kolom, 3))

for x in range(baris):
    for y in range(kolom):
        gmx = img[x,y] * alpa
        citra_kontras[x,y] = gmx

citra_kontras = citra_kontras.astype(np.uint8)

fig, axs = plt.subplots(2,2, figsize=(15,5))
axs[0,0].imshow(img)
axs[0, 0].set_title('Citra Asli')
axs[0,1].hist(img.ravel(),256,[0,256])
axs[0, 1].set_title('Histogram Citra Asli')
axs[1,0].imshow(citra_kontras)

axs[1, 0].set_title('Citra Kontras')
axs[1,1].hist(citra_kontras.ravel(),256,[0,256])
axs[1, 1].set_title('Histogram Citra Kontras')
plt.show()
```

## Operasi Meningkatkan Kecerahan

```python
beta = 30
citra_cerah = np.zeros((baris, kolom, 3))
for x in range(baris):
    for y in range(kolom):
        gyx = img[x,y] + beta
        citra_cerah[x,y] = gyx
        
citra_cerah = citra_cerah.astype(np.uint8)

fig, axs = plt.subplots(2,2, figsize=(15,5))
axs[0,0].imshow(img)
axs[0,1].hist(img.ravel(),256,[0,256])
axs[1,0].imshow(citra_cerah)
axs[1,1].hist(citra_cerah.ravel(),256,[0,256])
plt.show()
```

## Deteksi Warna Merah Pada Citra

```python
red_channel = citra_cerah[:,:,0]

plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.imshow(red_channel, cmap='gray')
plt.title('Red')

plt.subplot(1, 2, 2)
plt.hist(red_channel.ravel(), bins=256, alpha=0.7)
plt.xlabel('Intensity')
plt.ylabel('Frequency')
plt.title('Histogram Red')

plt.show()
```

## Deteksi Warna Hijau Pada Citra


```python
green_channel = citra_cerah[:,:,1]

plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.imshow(green_channel, cmap='gray')
plt.title('Green')

plt.subplot(1, 2, 2)
plt.hist(green_channel.ravel(), bins=256, alpha=0.7)
plt.xlabel('Intensity')
plt.ylabel('Frequency')
plt.title('Histogram Green')

plt.show()
```

## Deteksi Warna Biru Pada Citra


```python
blue_channel = citra_cerah[:,:,2]

plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.imshow(blue_channel, cmap='gray')
plt.title('Blue')

plt.subplot(1, 2, 2)
plt.hist(blue_channel.ravel(), bins=256, alpha=0.7)
plt.xlabel('Intensity')
plt.ylabel('Frequency')
plt.title('Histogram Blue')

plt.show()
```

## Nilai Ambang batas Citra Untuk Dapat Menampilan Kategori Warna Pada Citra

```python
hsv_image = cv2.cvtColor(citra_cerah, cv2.COLOR_RGB2HSV)

lower_blue = np.array([100, 100, 100])
upper_blue = np.array([140, 255, 255])

lower_green = np.array([20, 100, 100])
upper_green = np.array([250, 255, 255])

lower_red1 = np.array([0, 100, 100])
upper_red1 = np.array([10, 255, 255])
lower_red2 = np.array([160, 100, 100])
upper_red2 = np.array([180, 255, 255])

mask_blue = cv2.inRange(hsv_image, lower_blue, upper_blue)

mask_green = cv2.inRange(hsv_image, lower_green, upper_green)

mask_red1 = cv2.inRange(hsv_image, lower_red1, upper_red1)
mask_red2 = cv2.inRange(hsv_image, lower_red2, upper_red2)
mask_red = cv2.bitwise_or(mask_red1, mask_red2)


gray = cv2.cvtColor(citra_cerah, cv2.COLOR_RGB2GRAY)
fig, axs = plt.subplots (2, 2, figsize=(10,10))

(thresh, binary1) = cv2.threshold(gray, 0, 0, cv2.THRESH_BINARY)
axs[0,0].imshow(binary1, cmap = 'gray')
axs[0,0].set_title('NONE')

plt.subplot(2, 2, 2)
plt.imshow(mask_blue, cmap='gray')
plt.title('Blue')
plt.xticks(np.arange(0, mask_blue.shape[1]+1, 800))
plt.yticks(np.arange(0, mask_blue.shape[0]+1, 500))
plt.axis('on')

plt.subplot(2, 2, 3)
plt.imshow(np.maximum(mask_red, mask_blue), cmap='gray')
plt.title('red-blue')
plt.xticks(np.arange(0, mask_green.shape[1], 800))
plt.yticks(np.arange(0, mask_green.shape[0], 500))
plt.axis('on')

plt.subplot(2, 2, 4)
plt.imshow(mask_green, cmap='gray')
plt.title('Red-Green-Blue')
plt.xticks(np.arange(0, mask_green.shape[1], 800))
plt.yticks(np.arange(0, mask_green.shape[0], 500))
plt.axis('on')

plt.tight_layout()
plt.show()
```
